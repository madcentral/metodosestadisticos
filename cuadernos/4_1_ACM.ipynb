# Análisis de correspondencias múltiples
El ACM es una tecnica para reducir (mito urbano) dimensionalidad cuando se tienen variables cualitativas.
Es muy útil para volver un conjunto de datos cualitativo en cuantitativo y tambien para visualizar tablas de contingencia.

Un ACM trabaja sobre la tabla de frecuencias cruzadas, o tabla de Burt, la cual es una tabla simetrica con frecuencias cruzadas de cada variable categorica.

Sobre esta tabla se construyen valores y vectores propios, y además se contruye la distancia chi cuadrado, la cual se determina de la siguiente manera:

$$\chi_{col}^2=distancia(col\, i, col \, j)=\sum_{i=1}\frac{1}{} $$

```python
url="https://raw.githubusercontent.com/Cruzalirio/Ucentral/master/Bases/Diabetes.csv"
diabetes=pd.read_csv(url, sep=";", decimal=",")
diabetes.columns
```

```python
base=diabetes[['Gender', 'Polyuria', 'Polydipsia', 'sudden weight loss',
       'weakness', 'Polyphagia', 'Genital thrush', 'visual blurring',
       'Itching', 'Irritability', 'delayed healing', 'partial paresis',
       'muscle stiffness', 'Alopecia', 'Obesity', 'class']]
```

```python
acm=prince.MCA(n_components=20)
acm.fit(base)
nuevabase=acm.fit_transform(base)
nuevabase
```

```python
np.cumsum(acm.explained_inertia_)
```

```python
acm.plot_coordinates(base)
```
